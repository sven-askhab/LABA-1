Ответ: Количественные признаки (которые также называют вещественными или числовыми) могут быть дискретными или непрерывными.
Дискретные данные -принимают строго определенные значения. Например, это может быть количество сотрудников на предприятии или количество бракованных деталей в произведенной партии. Непрерывные данные - всегда выражены неограниченным числом значений.

2.Опишите разницу между обучением с учителем и обучением без учителя.

Ответ: Обучение с учителем — это подход, при котором машине заранее дают понять, какой ответ будет считаться правильным. Под учителем здесь понимается обучающая выборка с размеченными нужным образом данными либо тот, кто указывает правильные ответы в процессе обучения.

Обучение без учителя — это подход, который актуален ввиду большого количества неразмеченных, хаотично собранных данных. Машина должна сама найти связи между отдельными данными, выявить закономерности, подобрать шаблоны и отсортировать объекты в загруженном датасете.

3.Опишите разницу между задачами классификации и регрессии.

Ответ: Задача классификации в машинном обучении заключается в прогнозировании принадлежности объекта к определенному классу или категории. Например, определение, является ли изображение кошкой или собакой, или определение, к какому типу заболевания склонен пациент на основе медицинских данных.

С другой стороны, задача регрессии предполагает прогнозирование числового значения на основе входных данных. Например, предсказание цены дома на основе его характеристик, или прогнозирование времени, которое потребуется на выполнение определенной задачи.

Таким образом, основное различие между задачами классификации и регрессии заключается в том, что классификация предсказывает категорию, а регрессия - числовое значение. 4)Если вас попросят создать программу, которая будет определять кошку или собаку по изображению, к какому типу задач машинного обучения относится эта просьба.
